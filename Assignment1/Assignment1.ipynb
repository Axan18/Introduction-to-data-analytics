```python
#a)
#Your code:
import os
print(os.getcwd())

data = pd.read_csv("kc_house_data.csv")
data.head()

#Hint: read_csv(), head()
```

```python
#b) 

house_categorical_df = data[["id","date","waterfront","view","condition", "grade","zipcode"]]
#Categorical variables: id- not numerical, date- not numerical, waterfront- binary, view- ordinal, condition- ordinal, grade- ordinal, zipcode- nominal 
house_numerical_df = data[["price","bedrooms","sqft_living","sqft_lot","bathrooms","floors","sqft_above","sqft_basement","yr_built","yr_renovated","lat","long","sqft_living15","sqft_lot15"]]

```

```python
#c)
#Your code:

house_numerical_df.describe()
#Count is the same for all variables, so there are no missing values.
#For latitude and longitude, mean, 25%,50%,75% and standard deviation are not meaningful.
#For yr_renovated, 25%,50%,75%, mean and std are not meaningful because most of the values are 0.
#Hint: Function describe()
```

## Exercise 2
Using scatter plot:  
a) Show the relation of square footage and price.  
b) Check how price is influenced by apartment's grade. Find other variables correlated with price.  
c) Check how good apartments are distributed over the city. First, plot zipcode versus price. Then, use apartments coordinates (longitude and lattitude) in order to show where apartments of good grade are located. 

```python
#a)
#Your code:
plt.scatter(house_numerical_df["sqft_living"],house_numerical_df["price"])
plt.xlabel("living sqft")
plt.ylabel("price")

#Scatter plot of price and square footage of living area X-axis: sqft_living, Y-axis: price
#Hint plt.scatter
```

```python
#b)
#Your code:
plt.scatter(house_categorical_df["grade"],house_numerical_df["price"])
plt.xlabel("grade")
plt.ylabel("price")
plt.show()

plt.scatter(house_numerical_df["sqft_living"],house_numerical_df["price"])
plt.xlabel("living sqft")
plt.ylabel("price")
plt.show()

plt.scatter(house_numerical_df["sqft_above"],house_numerical_df["price"])
plt.xlabel("above sqft")
plt.ylabel("price")
plt.show()

plt.scatter(house_numerical_df["bathrooms"],house_numerical_df["price"])
plt.xlabel("bathrooms")
plt.ylabel("price")
plt.show()

plt.scatter(house_numerical_df["bedrooms"],house_numerical_df["price"])
plt.xlabel("bedrooms")
plt.ylabel("price")
plt.show()

#All the variables are somewhat correlated with price. The most correlated variables are sqft_living, sqft_above and grade.
#Scatter plot of price and grade X-axis: grade, Y-axis: price
```

```python
#c)
#Your code:

plt.scatter(house_categorical_df["zipcode"],house_numerical_df["price"])
plt.xlabel("zipcode")
plt.ylabel("price")
#We can see that price is influenced by zipcode. For some zipcodes, prices are higher than for others, but we don't know where these zipcodes are located.
```

```python
#Your code:
plt.scatter(house_numerical_df["long"],house_numerical_df["lat"], c=house_categorical_df["grade"])
plt.xlabel("longitude")
plt.ylabel("lattitude")
#There are some areas where apartments with higher grade are located, and it is much easier to see where they are located.
```

## Exercise 3
a) Create boxplot of apartment's grade versus it's price. Compare with plot 2b).  
b) Create mosaic plot of any two categorical variables.

```python
#a)
#Your code:
grouped_data = house_numerical_df.groupby(house_categorical_df["grade"])["price"].apply(list)
plt.boxplot(grouped_data)
plt.xlabel("grade")
plt.ylabel("price")

#Boxplot of price and grade X-axis: grade, Y-axis: price
#This plot is better than scatter plot from 2b) because we can see the distribution of prices for each grade. It gives us more information about the data.

#Hint: plt.boxplot
```

```python
#b)
from statsmodels.graphics.mosaicplot import mosaic

#Your code:
mosaic(house_numerical_df,["bedrooms","floors"])
plt.xlabel("bedrooms")
plt.ylabel("floors")
plt.show()

#Mosaic plot of bedrooms and floors
#This plot shows the distribution of apartments with different number of bedrooms and floors. Most of the apartments have 3 bedrooms and 1 floor.
```

## Exercise 4
"I am interested in the house with 3 bedrooms, 2 bathrooms and 2 floors"  
a) Compute the basic statistics of apartments fulfilling above constraints.  
b) Distinct accepted apartments on any of previously created plots.  
c) Find such apartments with best price to footage ratio.  

```python
#a)
#Your code:

filtered_data = data.loc[
    (data["bedrooms"] == 3) & 
    (data["bathrooms"] == 2) & 
    (data["floors"] == 2)
]
#Choose only the apartments with 3 bedrooms, 2 bathrooms and 2 floors
filtered_data.head()
#Hint: pandas.DataFrame.loc
```

```python
#b)
#Your code:
sc = plt.scatter(filtered_data["long"], filtered_data["lat"], c=filtered_data["price"], cmap='viridis')

plt.colorbar(sc, label='Price')

plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

```python
#c)
#Your code:
filtered_data.loc[:, "price_to_sqft"] = filtered_data["price"] / filtered_data["sqft_living"]
filtered_data.sort_values("price_to_sqft", ascending=True).head()

#Hint: .loc, .sort_values(), .index
```
